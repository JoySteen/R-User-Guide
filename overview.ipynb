## I. Plotting World Bank data collaboratively<br>with ggplot2, Python, R, and a web app

Suppose you see a lovely graph on the [World Bank website](http://blogs.worldbank.org/opendata/accessing-world-bank-data-apis-python-r-ruby-stata). The graph uses [ggplot2](http://ggplot2.org), a remarkable plotting library for R. You want to get the data from it, draw the graph with a log axis, and embed an interactive graph on your website.

We run the script in our RStudio (or the new [R kernel](https://github.com/takluyver/IRkernel) for IPython), add `py$ggplotly` to the call, and re-draw the ggplot2 figure in Plotly. It's drawn using D3, and adds the data to your graph at a URL: [plot.ly/RgraphingAPI/1457](https://plot.ly/RgraphingAPI/1457). 

```python
library(WDI)
library(ggplot2)
 
#Grab GNI per capita data for Chile, Hungary and Uruguay
 
dat = WDI(indicator='NY.GNP.PCAP.CD', country=c('CL','HU','UY'), start=1960, end=2012)
 
#a quick plot with legend, title and label
 
wb <- ggplot(dat, aes(year, NY.GNP.PCAP.CD, color=country)) + geom_line() 
+     xlab('Year') + ylab('GDI per capita (Atlas Method USD)') 
+     labs(title <- "GNI Per Capita ($USD Atlas Method)")

py$ggplotly(wb)
```

That produces an error, but that's ok: Plotly still graphs the data. Grab the response from https://plot.ly/~RgraphingAPI, make an iframe function, and add it to your iframe. 

```python
library(IRdisplay)
plotly_iframe <- function(url) {
    # set width and height from options or default square
    w <- "700"
    h <- "600"
    html <- paste("<center><iframe height=\"", h, "\" id=\"igraph\" scrolling=\"no\" seamless=\"seamless\"\n\t\t\t\tsrc=\"", 
        url, "\" width=\"", w, "\" frameBorder=\"0\"></iframe></center>", sep="")
    return(html)
}
```

```python
display_html(plotly_iframe("https://plot.ly/~RgraphingAPI/1645"))
```

## II. Other graph options from  ggplot2 and Plotly

You can use ggplot2 to make bar, line, and scatter charts. You can use [Plotly’s R API](plot.ly/r) to make those plus bubble, polar, contour, heatmaps, histograms, boxplots, 3D graphs, and streaming graphs. For example, here is a contour:

```python
library(plotly)
p <- plotly(username="MattSundquist", key="c6393zudwx")

size <- 100
x <- seq(-2*pi, 2*pi, length=size)
y <- seq(-2*pi, 2*pi, length=size)
z <- matrix(0, size, size)
for(i in 1:size) {
    for(j in 1:size) {
        r2 <- x[i]^2 + y[j]^2
        z[i, j] <- sin(x[i])*cos(y[j])*sin(r2)/log(r2+1)
    }
}

trace0 <- list(
  x = x, 
  y = y, 
  z = z, 
  type = "contour"
)
response <- p$plotly(trace0, kwargs=list(filename="simple-contour", fileopt="overwrite"))

url <- response$url
filename <- response$filename
```

## III. Bar Charts.

Let's take on a [grouped bar chart](http://stackoverflow.com/questions/18158461/grouped-bar-plot-in-ggplot) question from Stack Overflow. We'll need to use a few extra packages.

```python
install.packages("reshape", quiet = TRUE)
```

```python
library(reshape, quiet = TRUE)
```

And read in our data.

```python
raw <- read.csv("http://pastebin.com/raw.php?i=L8cEKcxS",sep=",")
raw[,2]<-factor(raw[,2],levels=c("Very Bad","Bad","Good","Very Good"),ordered=FALSE)
raw[,3]<-factor(raw[,3],levels=c("Very Bad","Bad","Good","Very Good"),ordered=FALSE)
raw[,4]<-factor(raw[,4],levels=c("Very Bad","Bad","Good","Very Good"),ordered=FALSE)

raw=raw[,c(2,3,4)] # getting rid of the "people" variable as I see no use for it

freq=table(col(raw), as.matrix(raw)) # get the counts of each factor level
```

```python
Names=c("Food","Music","People")     # create list of names
data=data.frame(cbind(freq),Names)   # combine them into a data frame
data=data[,c(5,3,1,2,4)]             # sort columns

# melt the data frame for plotting
data.m <- melt(data, id.vars='Names')

# plot everything
n <- ggplot(data.m, aes(Names, value)) +   
  geom_bar(aes(fill = variable), position = "dodge", stat="identity")
  py$ggplotly(n)
```

Let's try a [stacked bar charts](http://stackoverflow.com/questions/21236229/ggplot2-stacked-bar-chart).

```python
DF <- read.table(text="Rank F1     F2     F3
1    500    250    50
2    400    100    30
3    300    155    100
4    200    90     10", header=TRUE)

library(reshape2)
DF1 <- melt(DF, id.var="Rank")

library(ggplot2)
b <- ggplot(DF1, aes(x = Rank, y = value, fill = variable)) + 
  geom_bar(stat = "identity")
py$ggplotly(b)
```

And we can do a [dodged bar chart](http://stackoverflow.com/questions/14587335/how-to-ggplot-dodged-bar-chart-using-stat-identity)

```python
d <- data.frame(A=c(1:10), B=c(11:20))
ggplot(data=data.frame(value=c(d$A, d$B), variable=c(rep("A",10),rep("B",10))), aes(x=c(1:10,1:10), y=value, fill=variable))+geom_bar(stat="identity", position="dodge")
py$ggplotly()
```
